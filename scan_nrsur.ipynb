```python
import gwsurrogate
import gwpy
import matplotlib.pyplot as plt
import matplotlib
from gwpy.timeseries import TimeSeries
import matplotlib.colors as colors
import numpy as np
```

```python
sur=gwsurrogate.LoadSurrogate('NRSur7dq4')
```

```python
q = 4
chiA = [0,0,0]
chiB = [0,0,0]
M = 44
dist_mpc = 100
ellMax = 4     
inclination = np.pi/2
f_low=0
dt=1/4096

t, h, dyn = sur(q, chiA, chiB, dt=dt, f_low=f_low, ellMax=ellMax, M=M, dist_mpc=dist_mpc,
               inclination=inclination, units='mks')
plt.plot(t, h)
plt.xlim(-0.6,0.01)
```

```python
hdata = TimeSeries(h, times=t)
```

```python
hq = hdata.q_transform(whiten=False, outseg=(-0.5,0.01))
hq=np.abs(hq)
hq.max()
```

```python
plot = hq.plot()
plot.colorbar(label="Normalised energy", vmin=1e3, vmax=9e9, norm='log')
ax = plot.gca()
ax.set_yscale('log')
```
